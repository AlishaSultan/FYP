```python
import os
import rasterio
from PIL import Image
import numpy as np

# Replace 'dataset_root_folder' and 'dataset_name' with the actual path and name of your dataset
dataset_root_folder = 'Data/'
dataset_name = "DubaiDataset"

for root, dirs, files in os.walk(os.path.join(dataset_root_folder, dataset_name)):
    for dir_name in dirs:
        # Check if the directory is an 'images' directory
        if dir_name.lower() == 'images':
            # Construct the full path to the 'images' folder
            images_folder = os.path.join(root, dir_name)

            # Iterate through files in the 'images' folder
            for image_file in os.listdir(images_folder):
                # Construct the full path to the image
                image_path = os.path.join(images_folder, image_file)

                # Open the RGB satellite image using rasterio
                with rasterio.open(image_path) as src:
                    # Read the RGB bands (assuming they are the first three bands)
                    rgb_bands = src.read([1, 2, 3])

                    # Convert the band data to uint8 (required for JPEG)
                    scaled_band_data = (rgb_bands / rgb_bands.max() * 255).astype('uint8')
                    

                    # Create an RGB image using the three bands
                    rgb_image = Image.fromarray(scaled_band_data.transpose(1, 2, 0))

                    # Print the shape of the NumPy array representation of the image
                    image_array = np.array(rgb_image)
                    print(f"Shape of the RGB image: {image_array.shape}")

                    # Save the RGB image to a JPEG file
                    output_folder = 'Data/DubaiDataset/Output'
                    output_filename = f"{os.path.splitext(image_file)[0]}_rgb.jpg"
                    output_path = os.path.join(output_folder, output_filename)
                    rgb_image.save(output_path)
```
